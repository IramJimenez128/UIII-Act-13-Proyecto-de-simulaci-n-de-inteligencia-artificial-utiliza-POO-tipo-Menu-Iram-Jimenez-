#### Problema sobregiro bancario
### Atributos Credito_automatico, saldo-negativo, Costos-elevados, Temporalidad, Origen
#### Funciones Cobertura_Pagos(), Financiación_Inmediata(), Gestion_Liquidez_Temporal(), Evitar_Penalizaciones(), Conectividad_Cuentas()

```python
# --- SISTEMA DE IA BANCARIA: MÓDULO DE SOBREGIRO V1.0 ---

class SobregiroBancario:
    """Clase que representa la lógica de IA para gestionar sobregiros."""

    def __init__(self, credito, saldo, costos, tiempo, origen):
        """Constructor que inicializa los 5 atributos del sobregiro [5]."""
        self.Credito_automatico = credito
        self.saldo_negativo = saldo
        self.Costos_elevados = costos
        self.Temporalidad = tiempo
        self.Origen = origen

    # --- FUNCIONES DE LÓGICA DE IA (MÉTODOS) ---

    def Cobertura_Pagos(self):
        """Analiza si el crédito automático cubre el saldo negativo [6]."""
        if self.Credito_automatico >= abs(self.saldo_negativo):
            return "IA: COBERTURA TOTAL. Pagos garantizados."
        else:
            return "IA: ALERTA. Crédito insuficiente para cubrir deuda."

    def Financiacion_Inmediata(self):
        """Evalúa la viabilidad de financiamiento rápido [7]."""
        if self.saldo_negativo > -5000:
            return "APROBADO: Financiamiento disponible al instante."
        else:
            return "RECHAZADO: Requiere evaluación de riesgo manual."

    def Gestion_Liquidez_Temporal(self):
        """Determina la salud financiera según la temporalidad del sobregiro."""
        if self.Temporalidad == "corto":
            return "ESTADO: Liquidez estable. Riesgo mínimo."
        else:
            return "ESTADO: Liquidez crítica. Posible insolvencia."

    def Evitar_Penalizaciones(self):
        """Brinda consejos según el nivel de costos elevados [8]."""
        if self.Costos_elevados == "si":
            return "CONSEJO: Cubra el saldo en < 24hrs para evitar cargos."
        else:
            return "CONSEJO: Mantenga el monitoreo, costos dentro de margen."

    def Conectividad_Cuentas(self):
        """Analiza el origen de los fondos para el respaldo."""
        return f"ORIGEN DETECTADO: {self.Origen.capitalize()}. Conectando cuentas..."

# --- LÓGICA DEL PROGRAMA PRINCIPAL (MENÚ INTERACTIVO) ---

# Lista para almacenar los objetos creados [9]
registros_sobregiro = []

while True:
    print("\n" + "="*50)
    print("      SISTEMA INTELIGENTE DE SOBREGIRO (AI)")
    print("="*50)
    print("1. Registrar nuevo sobregiro")
    print("2. Listado de registros (Tabla)")
    print("3. Ejecutar diagnóstico de IA")
    print("4. Salir")

    opcion = input("\nAI: Seleccione una acción (1-4): ")

    if opcion == "1":
        print("\n--- Entrada de Datos del Cliente ---")
        credito = float(input("¿Monto de Crédito Automático?: "))
        saldo = float(input("¿Monto del Saldo Negativo? (ej: -1500): "))
        costos = input("¿Existen Costos Elevados? (si/no): ").lower()
        tiempo = input("¿Temporalidad? (corto/largo): ").lower()
        origen = input("¿Origen de la cuenta? (ahorros/nomina): ").lower()

        # Instanciación: Crear el objeto y guardarlo en la lista [10, 11]
        nuevo_caso = SobregiroBancario(credito, saldo, costos, tiempo, origen)
        registros_sobregiro.append(nuevo_caso)
        print("\nIA: Datos procesados y almacenados con éxito.")

    elif opcion == "2":
        if not registros_sobregiro:
            print("\nIA: La base de datos está vacía.")
        else:
            # Encabezado de tabla con alineación perfecta [3]
            header = f"{'ID':<3} | {'Crédito':<10} | {'Saldo Neg.':<12} | {'Costos':<8} | {'Tiempo':<8} | {'Origen':<10}"
            print("\n" + "-"*65)
            print(header)
            print("-"*65)

            # Recorrido de objetos en la lista [12]
            contador = 1
            for caso in registros_sobregiro:
                fila = f"{contador:<3} | {caso.Credito_automatico:<10.2f} | {caso.saldo_negativo:<12.2f} | {caso.Costos_elevados:<8} | {caso.Temporalidad:<8} | {caso.Origen:<10}"
                print(fila)
                contador += 1
            print("-"*65)

    elif opcion == "3":
        if not registros_sobregiro:
            print("\nIA: No hay casos para diagnosticar.")
        else:
            print("\n--- INFORME DE ANÁLISIS DE INTELIGENCIA ARTIFICIAL ---")
            for i, caso in enumerate(registros_sobregiro, 1):
                print(f"\n>> ANÁLISIS DEL CASO #{i} (Saldo: {caso.saldo_negativo})")
                # Llamada a las funciones de la clase [13, 14]
                print(f"- {caso.Cobertura_Pagos()}")
                print(f"- {caso.Financiacion_Inmediata()}")
                print(f"- {caso.Gestion_Liquidez_Temporal()}")
                print(f"- {caso.Evitar_Penalizaciones()}")
                print(f"- {caso.Conectividad_Cuentas()}")

    elif opcion == "4":
        print("\nIA: Desconectando sistemas de seguridad... ¡Hasta pronto!")
        break # Rompe el ciclo para finalizar el programa [15]

    else:
        print("\nIA: Opción no válida. Por favor, reintente.")

print("\n--- SESIÓN FINALIZADA ---")
print("rograma realizado por Iram Jimenez 2H y con ayuda de ChatGPT y Notebooks")
```
